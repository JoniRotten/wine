```python
#normalitzem
from sklearn.preprocessing import StandardScaler
autoscaler = StandardScaler()
data_scaler = autoscaler.fit_transform(data)
#traiem la columna de les categories
data_norm = pd.DataFrame(data= data_scaler[:,0:13], columns= wine['feature_names'])
```

```python
#pca, reduim dimensió
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(data_norm)
#Creem el DataFrame amb les dues components i les etiquetes
data_PCA = pd.DataFrame(data= np.c_[principalComponents, wine['target']],
                     columns= ['principal_component_1' , 'principal_component_2' , 'target'])
```

```python
#Generem la gràfica
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = [0, 1, 2]
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = data_PCA['target'] == target
    ax.scatter(data_PCA.loc[indicesToKeep, 'principal_component_1']
               , data_PCA.loc[indicesToKeep, 'principal_component_2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
#Apliquem TSNE sense perplexity
tsne = TSNE(n_components=2).fit_transform(data_norm)
data_TSNE = pd.DataFrame(data= np.c_[tsne, wine['target']],
                     columns= ['principal_component_1' , 'principal_component_2' , 'target'])
```

```python
#Generem la gràfica
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component TSNE', fontsize = 20)
targets = [0, 1, 2]
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = data_TSNE['target'] == target
    ax.scatter(data_TSNE.loc[indicesToKeep, 'principal_component_1']
               , data_TSNE.loc[indicesToKeep, 'principal_component_2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
